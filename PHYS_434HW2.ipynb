1) A little introductory brain teaser. Which is more probable when rolling 2 six-sided dice: rolling snake eyes (two ones) or rolling sevens (dice sum to seven)? What is the ratio of the probabilities?

Rolling two one's should be (1/6)^2, cooresponding to a probability of $\frac{1}{36}$. Rolling two dice that sum to seven can be a combination of 1+6, 2+5, 3+4, 4+3, 5+2, 6+1 so the odds would be $\frac{6}{36}$, which is 6 times the probability of rolling snake eyes.
